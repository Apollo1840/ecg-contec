```python
%%capture
import os
import numpy as np
import matplotlib.pyplot as plt

from ecgAI import ECG, CINC2020_CPSC
from ecg2scp import ECGBasic

data_dir = "/home/congyu/congyu_program/pythons/forks/ecg-contec/examples/"

```

# Case I

```python
# load from scp
rid = "tmp1"
ecg = ECG.from_scp("test", rid, data_dir=data_dir)
plt.figure(figsize=(30, 5))
plt.plot(ecg.data[0])
plt.show()


# export scp file
ecg_b = ECGBasic(np.array(ecg.data*5000, dtype=np.int32))
plt.figure(figsize=(30, 5))
plt.plot(ecg_b.data[0])
plt.show()
print(ecg_b.data.shape)

ecg_b.export_scp(os.path.join(data_dir, "{}_r.scp".format(rid)))

# reload from ther exported scp
ecg_r = ECG.from_scp("test", "{}_r".format(rid), data_dir=data_dir)
plt.figure(figsize=(30, 5))
plt.plot(ecg_r.data[0])
plt.show()
print(ecg_r.data.shape)
```
