```python
%%capture
import numpy as np
import matplotlib.pyplot as plt

from ecgAI import ECG, CINC2020_CPSC
from ecg2scp import ECGBasic

data_dir = "/home/congyu/congyu_program/pythons/forks/ecg-contec/examples/"
```

# load from scp

```python
ecg = ECG.from_scp("test", "tmp1", data_dir=data_dir)
plt.plot(ecg.data[0])
```

# create scp from np.array

```python
ecg2 = ECGBasic()
ecg2.data = np.array(ecg.data*5000, dtype=np.int32)
plt.plot(ecg2.data[0])
```

```python
import sys
import struct

val = ecg2.data[0, 0]
print(val)
print(type(val))
print(sys.getsizeof(val), "bits")
# 1 byte == 8 bits
print(sys.getsizeof(val)//8, "bytes")
b = struct.pack('<h', val)
print(b)
print(len(b), "bytes")
print("->", int.from_bytes(b, "little"))

print("-"*32)

val_int = int(val)
print(val_int)
print(type(val_int))
print(sys.getsizeof(val_int), "bits")
# 1 byte == 8 bits
print(sys.getsizeof(val_int)//8, "bytes")
b = struct.pack('<h', val_int)
print(b)
print(len(b), "bytes")
print("->", int.from_bytes(b, "little"))
```

```python
ecg2.export_scp("examples/tmp1_r.scp")
```

# reload the scp

```python
ecg_r = ECG.from_scp("test", "tmp1_r", data_dir=data_dir)
plt.plot(ecg_r.data[0])
```
